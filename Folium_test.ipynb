```python
import folium
from folium import plugins
from branca.element import Figure
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
fig=Figure( width=600, height=400)
map=folium.Map()
```

```python
Lyon = folium.Map(location=[4.7640,4.8357],zoom_start=7, width=600, height=400)
fig=Figure( width=600, height=400)

fig.add_child(Lyon)
Lyon 
```

```python
#enregistrer le carte
Lyon.save('/usr/lyon.html')
```

```python
#ajouter divers couches
folium.TileLayer('Stamen Water Color').add_to(Lyon)
folium.TileLayer('Stamen Terrain').add_to(Lyon)
folium.TileLayer('Stamen Toner').add_to(Lyon)
folium.LayerControl().add_to(Lyon)

Lyon
```

```python
folium.Marker(location=[45.7640,4.8357],popup='Lyon').add_to(Lyon)
folium.CircleMarker([45.7640, 4.8357],
                    radius=50,
                    popup='Lyon',
                    color='red',
                    ).add_to(Lyon)
Lyon
```

```python
france_cities_df = pd.DataFrame({'city': ['Paris', 'Marseille', 'Lyon'],
                           'latitude': [48.8566, 43.2965, 45.7640],
                           'longitude': [2.3522, 5.3698, 4.8357],
                           'icon_num': [1, 2, 3]})

fig=Figure( width=600, height=400)
map_france_icons = folium.Map([45, 4], zoom_start=5)

# icons using plugins.BeautifyIcon
for i in france_cities_df.itertuples():
    folium.Marker(location=[i.latitude, i.longitude],
                  popup=i.city,
                  icon=plugins.BeautifyIcon(number=i.icon_num,
                                            border_color='blue',
                                            border_width=1,
                                            text_color='red',
                                            inner_icon_style='margin-top:0px;')).add_to(map_france_icons)

fig.add_child(map_france_icons)
map_france_icons
```
